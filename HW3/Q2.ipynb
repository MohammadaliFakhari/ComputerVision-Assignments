```python
import numpy as np
import cv2
import glob
import matplotlib.pylab as plt
import os
from google.colab.patches import cv2_imshow
```

<h3><font color = "#079e05"> Read input images from 1 to 8 and show them in a row together </font></h3>

```python
path = '/content/drive/MyDrive/ColabNotebooks/FCV/HW3/images/Q2/'
images = []
for i in range(1, 9):
  img = cv2.imread(os.path.join(path, f'{i}.jpg'))
  images.append(cv2.resize(img,(0,0),fx=0.5,fy=0.5))

combined = cv2.hconcat(images)
cv2_imshow(combined)
```

<h3><font color = "#079e05"> Initialize OpenCV's image sticher object and then perform the image stitching on input images </font></h3>

```python
sticher_obj = cv2.Stitcher().create()
(status, output) = sticher_obj.stitch(images)
```

<h3><font color = "#079e05"> Show the result </font></h3>

```python
if status != cv2.STITCHER_OK:
  print('somethimg wents wrong.')
else:
  print('done.')

cv2_imshow(output)
```
